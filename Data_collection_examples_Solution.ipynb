## Task C

Analyze the content of the Point traffic sensor data. 



Data is downloaded from : https://www.muoversiatorino.it/it/open-data/

Check FDT schema (in italian): https://simone.5t.torino.it/simone-definizione_protocollo_comunicazione_400.pdf

- speed - average speed of flow in [km/h]
- flow - measured as number of vehicles [Vehicles/h] 

```python
import pandas_read_xml as pdx
df = pdx.read_xml("Dataset2.xml", ['traffic_data','FDT_data'],root_is_rows=False)
df=pdx.flatten(df)
df
```

    • Which is the road with the highest flow (vehicles/h)?


```python
df['speedflow|@flow']=pd.to_numeric(df['speedflow|@flow'])
max_flow=df['speedflow|@flow'].max()
print("max flow is : ", max_flow, "vehicles per hour")
```

```python
max_flow_streets=df[df['speedflow|@flow']==max_flow]
max_flow_streets
```

```python
print("Road(s) with highest flow: ", [el for el in max_flow_streets['@Road_name']])
```

    • Which is the lowest average flow speed (km/h)?


```python
#we do not consider sensors with null flow

df['speedflow|@speed']=pd.to_numeric(df['speedflow|@speed'])
min_speed=df[df['speedflow|@flow']>0]['speedflow|@speed'].min()
print("Min speed is : ", min_speed, " km/h")
```

```python
#Since 0 is not likely (if there is a flow), let's consider only positive speed
min_speed=df[df['speedflow|@speed']>0]['speedflow|@speed'].min()
print("Min non-zero speed is : ", min_speed, " km/h")
```

    • Locate through the coordinates in a map the position of the inductive loop further to the east. 


Further to the east means with largest longitude (EAST)

```python
df['@lng']=pd.to_numeric(df['@lng'])
east_lon=df['@lng'].max()
print("Largest lon : ", east_lon)
```

```python
east_loc=df[df['@lng']==east_lon]
east_loc
```

```python
import geopandas as gpd
from geopandas import GeoDataFrame

geometry = gpd.points_from_xy(df["@lng"],df["@lat"],crs="EPSG:4326")
gdf = GeoDataFrame(df, geometry=geometry)   
gdf.explore()
```

```python
import geopandas as gpd
from geopandas import GeoDataFrame

geometry = gpd.points_from_xy(east_loc["@lng"],east_loc["@lat"],crs="EPSG:4326")
gdf = GeoDataFrame(east_loc, geometry=geometry)   
gdf.explore()
```

    • Where is located the stop with latest arrival time during the day (arrival_time)?

for arrival times the table to check is stop_times.txt:  https://developers.google.com/transit/gtfs/reference#stop_timestxt

```python
df_stopTimes = pd.read_csv ('Dataset1/stop_times.txt')
df_stopTimes
```

Consider latest arrival time during the 24 hours (i.e., arrival time <= 23:59:59)

```python
df_stopTimes["arrival_time_day"]=pd.to_datetime(df_stopTimes["arrival_time"].replace({'24:': '00:'}, regex=True).replace({'25:': '00:'}, regex=True), format='%H:%M:%S').dt.time
```

```python
latest_arrival_time_day=df_stopTimes["arrival_time_day"].max()
print("Latest arrival during the 24 hours: ", latest_arrival_time_day)
```

```python
df_latest=df_stopTimes[df_stopTimes["arrival_time_day"]==latest_arrival_time_day]
df_latest
```

```python
print("Stops related to latest arrival during the 24 hours: ", [el for el in df_latest["stop_id"]])
```

```python
stop_target=[el for el in df_latest["stop_id"]][0]
df_stop_target=df_stops[df_stops["stop_id"]==stop_target]
df_stop_target
```

where are stops located

```python
import geopandas as gpd
from geopandas import GeoDataFrame

geometry = gpd.points_from_xy(df_stops["stop_lon"],df_stops["stop_lat"],crs="EPSG:4326")
gdf = GeoDataFrame(df_stops, geometry=geometry)   
gdf.explore()
```

```python
geometry = gpd.points_from_xy(df_stop_target["stop_lon"],df_stop_target["stop_lat"],crs="EPSG:4326")
gdf = GeoDataFrame(df_stop_target, geometry=geometry)   
gdf.explore()
```

Alternatively, we can repeat the exercise considering latest arrival time considering timetable (can be the day after, i.e., arrival time>=24:00:00)

```python
df_stopTimes["arrival_time_total_time"]=pd.to_timedelta(df_stopTimes["arrival_time"])
df_stopTimes
```

```python
latest_arrival_time=df_stopTimes["arrival_time_total_time"].max()
print("Latest arrival: ", latest_arrival_time)
```

```python
df_latest_total=df_stopTimes[df_stopTimes["arrival_time_total_time"]==latest_arrival_time]
df_latest_total
```
